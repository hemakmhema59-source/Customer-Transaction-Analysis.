```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load data (WORKING dataset)
url = "https://raw.githubusercontent.com/sharmaroshan/Churn-Modelling-Dataset/master/Churn_Modelling.csv"
data = pd.read_csv(url)

print(data.head())

# 2. Data cleaning
data = data.dropna().drop_duplicates()

# 3. Basic analysis
print("\nExited (Churn) Count:\n", data['Exited'].value_counts())

# 4. Simple anomaly detection (high balance)
threshold = data['Balance'].mean() + 2 * data['Balance'].std()
data['Anomaly'] = data['Balance'] > threshold

print("\nAnomalies:\n", data['Anomaly'].value_counts())

# 5. Visualization
plt.figure()
data['Exited'].value_counts().plot(kind='bar')
plt.title("Customer Churn")
plt.show()

# 6. Insights
print("\nInsights:")
print("Few customers leave the bank.")
print("High balance customers may show unusual patterns.")
```
